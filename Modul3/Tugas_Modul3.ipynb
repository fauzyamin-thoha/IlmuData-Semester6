```python
# Supervised Learning - Regresi Linear
from sklearn.linear_model import LinearRegression
import numpy as np
X = np.random.rand(100, 1)
y = 3 * X.squeeze() + np.random.randn(100)
model = LinearRegression()
model.fit(X, y)
print("Koefisien:", model.coef_)
```
